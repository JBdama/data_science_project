```python
import pandas as pd
import matplotlib.pyplot as plt

test_countries = ['USA', 'Canada', 'Mexico', 'Germany', 'France', 'UK', 'Japan', 'China', 'Brazil', 'Argentina']

dataframe = pd.read_csv('data.csv')


# test_dataframe = dataframe[dataframe['Country'].isin(test_countries)]
test_dataframe = dataframe.iloc[:10]
print("Filtered frame: ")
print(test_dataframe)
# for country in test_countries:
feature_1 = 'SafetySecurity'
feature_2 = 'PersonelFreedom'
plt.plot(test_countries, test_dataframe[feature_1], label=feature_1, color="blue")

plt.plot(test_countries, test_dataframe[feature_2], label=feature_2, color="red")

difference = test_dataframe[feature_1] - test_dataframe[feature_2]
difference_mean = difference.mean()
print("Mean difference: ", difference_mean)
plt.plot(test_countries, difference, label="Difference", color="green")

plt.show()
```
